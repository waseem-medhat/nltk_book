## 7. ☼ According to Strunk and White's Elements of Style, the word however, used at the start of a sentence, means "in whatever way" or "to whatever extent", and not "nevertheless". They give this example of correct usage: However you advise him, he will probably do as he thinks best. (http://www.bartleby.com/141/strunk3.html) Use the concordance tool to study actual usage of this word in the various texts we have been considering.

```python
def however_style(nltk_text):
    """
    Counts the forms of "however" being used in a given text object.
    If a form is not like the two main ones, some context will print.
    """

    conc_however = nltk_text.concordance_list("however", 50, len(nltk_text))

    however_counts = {
        ", however,": 0,
        "However,": 0,
        "other": 0
    }

    for ci in conc_however:

        lft_token = ci[0][-1]
        rgt_token = ci[2][0]

        if lft_token == "," and rgt_token == ",":
            however_counts[", however,"] += 1
        elif lft_token == "." and rgt_token == ",":
            however_counts["However,"] +=1
        else:
            print(
                " ".join(ci[0][-3:]),
                ci[1],
                " ".join(ci[2][:3])
            )
            however_counts["other"] +=1

    return however_counts
```

```python
text1 = nltk.Text(gutenberg.words(fileids="austen-persuasion.txt"))
however_style(text1)
```

```python
text2 = nltk.Text(gutenberg.words("melville-moby_dick.txt"))
however_style(text2)
```

## 9. ◑ Pick a pair of texts and study the differences between them, in terms of vocabulary, vocabulary richness, genre, etc. Can you find pairs of words which have quite different meanings across the two texts, such as monstrous in Moby Dick and in Sense and Sensibility?

```python
def vocab_richness(words):
    vocab = set([w.lower() for w in words])
    vrich = len(vocab) / len(words)
    return {
        "vocabulary": len(vocab),
        "words": len(words),
        "richness" : str(round(100*vrich, 2)) + "%"
    }
```

```python
def common_content_words(words):

    sw_en = stopwords.words("english")
    content_words = [
        w for w in words 
        if not w.lower() in sw_en 
        and w.isalpha()
    ]
    
    return nltk.FreqDist(content_words).most_common(10)
```

```python
for book in ["bible-kjv.txt", "shakespeare-hamlet.txt"]:
    
    words = gutenberg.words(book)
    title_slice = words[ words.index("[")+1 : words.index("]") ]
    
    print(
        ' '.join(title_slice),
        vocab_richness(words),
        "\nCommon words:",
        common_content_words(words),
        
        sep = "\n"
    )
    
    print("\n")
```

- I tried to filter out the stop words using the English list provided by NLTK. But that list does not include the arcahic words, so the frequency distributions include words like "thou" and "thy".
- The word "lord" seems commonly occurring in both texts. We can examine if they hold the same meaning.

```python
nltk.Text(gutenberg.words("bible-kjv.txt")) \
    .similar("LORD")
```

## 10. ◑ Read the BBC News article: UK's Vicky Pollards 'left behind' http://news.bbc.co.uk/1/hi/education/6173441.stm. The article gives the following statistic about teen language: "the top 20 words used, including yeah, no, but and like, account for around a third of all words." How many word types account for a third of all word tokens, for a variety of text sources? What do you conclude about this statistic? Read more about this on LanguageLog, at http://itre.cis.upenn.edu/~myl/languagelog/archives/003993.html.

```python
def top_third(words):
    """
    Takes a list of words (e.g. from a corpus).
    Returns the smallest number of the most common words such that
    their frequencies sum up to a third of the whole text length.
    """
    
    fd = nltk.FreqDist([w for w in words if w.isalpha()])
    freqs = sorted(dict(fd).values(), reverse=True)

    for i,_ in enumerate(freqs):
        if sum(freqs[:i]) / sum(freqs) > 1/3:
            return i
```

```python
list(map(
    lambda x: top_third(gutenberg.words(x)),
    gutenberg.fileids()
))
```

```python
list(map(
    lambda x: top_third(brown.words(categories=x)),
    brown.categories()
))
```

- To clarify the idea about that statistic, if a higher number of the most common words is needed to account for a third of the whole text, then this text supposedly has richer vocabulary and doesn't use a small set of words too frequently.

- After calculating the same statistic for different texts from the Gutenberg corpus (by book) and Brown corpus (by genre), it seems that the number 20 is not that bad. Actually, it is quite typical to have that number in the 20s range. It even was 17 in one of the cases, which is the King James bible.

```python
top_third(gutenberg.words("bible-kjv.txt"))
```

## 11. ◑ Investigate the table of modal distributions and look for other patterns. Try to explain them in terms of your own impressionistic understanding of the different genres. Can you find other closed classes of words that exhibit significant differences across different genres?

```python
def modal_freq(genre):
    """
    Takes a string representing a genre in the Brown Corpus.
    Returns a plot of the frequency distribution of modal verbs.
    """

    modals = "can could may might must will".split()
    
    return nltk.FreqDist(
        [m for m in brown.words(categories=genre)if m in modals]
    )
```

```python
list(map(lambda g: modal_freq(g).plot(title = g.title()), brown.categories()))
```

```python
list(map(lambda g: (g, dict(modal_freq(g))), brown.categories()))
```

In general, the most common modal in any genre is one of "can", "could" or "will", and the least common is either "may" or "might" (with the exception of reviews having "must" as the least common modal). The following points are some explanations on why these findings make sense.
- Genres like _mystery_, _fiction_, _adventure_, and _romance_ typically involve uncertainty about future events which is why a word like "could" is the most common.
- Modals in _news_ are dominated by the word "will". It makes sense to think that news are not only about current and past events (which are not associated with modals) but also future events.
- _Learned_ and _hobbies_ genres include educational/informational content that tell their readers what the "can" do.

## 14. ◑ Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s. 

```python
def supergloss(s):
    """
    Takes a synset (s)
    Returns the definition of itself, its hypernyms and 
    its hyponyms, all concatenated in a single string.
    """
    
    def_s     = [s.definition()]
    def_hyper = list(map(lambda s: s.definition(), s.hypernyms()))
    def_hypo  = list(map(lambda s: s.definition(), s.hyponyms()))
    
    return ', '.join(def_s + def_hyper + def_hypo)
```

## 17.◑ Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

Note: the exercise does not specify that puncuation should be also excluded. But I will do it in my implementation because I think it makes more sense than getting commas as the most common "word".

```python
def most_common_50(word_list):
    """
    Takes a word list (e.g. from a corpus)
    Returns the 50 most frequent words, 
    excluding stopwords and punctuation
    """
    
    non_stop = [w for w in word_list
                if w.isalpha()
                and not w.lower() in stopwords.words("english")]
    fd = nltk.FreqDist(non_stop)
    
    return fd.most_common(50)
```

```python
most_common_50(brown.words(categories="news"))
```

## 18. ◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
def most_common_bigrams(word_list):
    """
    Takes a word list (e.g. from a corpus)
    Returns the 50 most common bigrams,
    excluding stopwords and punctuation
    """
    
    non_stop = [w for w in word_list
            if w.isalpha()
            and not w.lower() in stopwords.words("english")]
    fd = nltk.FreqDist(
        list(nltk.bigrams(non_stop))
    )
    
    return fd.most_common(50)
```

```python
most_common_bigrams(brown.words(categories="news"))
```

## 19. ◑ Write a program to create a table of word frequencies by genre, like the one given in 1 for modals. Choose your own words and try to find words whose presence (or absence) is typical of a genre. Discuss your findings.

```python
def freq_in_genres(word_list):
    """
    Takes any word list
    Returns the frequency of each word in each genre of
    the Brown Corpus
    """
    
    cfd = nltk.ConditionalFreqDist(
        (genre, word)
        for word in word_list
        for genre in brown.categories()
        for token in brown.words(categories=genre)
        if token.lower() == word.lower()
    )
    
    return cfd.tabulate()
```

```python
freq_in_genres(["died", "legislation", "atom"])
```

Most frequencies make sense, except that the word "atom" is unexpecedly most common in religion texts. We can investigate some contexts.

## 20. ◑ Write a function word_freq() that takes a word and the name of a section of the Brown Corpus as arguments, and computes the frequency of the word in that section of the corpus.

```python
def word_freq(word, section):
    """
    Takes two strings: a word and a section name of the Brown Corpus
    (e.g. "news")
    Returns the frequency of the word in texts of the given section
    """
    
    section_words = [w.lower() for w in brown.words(categories=section)]
    freqs = dict(nltk.FreqDist(section_words))
    
    if word.lower() in freqs.keys():
        return freqs[word.lower()]
    else:
        return 0
```

```python
word_freq("atom", "religion")
```

## 21. ◑ Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary.

```python
def count_syllables(word_list):
    """
    Takes any word list
    Returns a count the number of syllables of the whole text
    based on the CMU Pronouncing Dictionary
    """

    def num_syllables(plist):
        """
        Takes a list of ARPABET pronunciations
        Returns the number of found digits which correspond to syllables
        """
        stress = [re.findall("\d", p) for p in plist]
        digits = [l[0] for l in stress if not l == []]

        return len(digits)
    
    indices   = [cmudict.words().index(w.lower()) for w in word_list]
    prons     = [cmudict.entries()[i][1] for i in indices]     
    syllables = [num_syllables(plist) for plist in prons]
    
    return sum(syllables)
```

```python
test = "You could be my unintended choice to live my life extended you could be the one".split()
count_syllables(test)
```


## 22. ◑ Define a function hedge(text) which processes a text and produces a new version with the word 'like' between every third word.

```python
def hedge(text):
    """
    Takes a word list
    Returns the list with "like" added after every third word
    """
    
    def triples(text):
        """
        Takes a word list
        Returns a nested list where each element is a 3-word list
        """

        triples = []

        for i,_ in enumerate(text):

            trigger = i%3

            if trigger == 0:

                remaining = len(text) - i

                if   remaining == 0: break
                elif remaining == 1: triples.append([text[i]])
                elif remaining == 2: triples.append([text[i], text[i+1]])
                else:
                    triples.append([text[i], text[i+1], text[i+2]])

        return triples
    
    hedged_triple = [wl + ["like"] for wl in triples(text)]
    hedged = reduce(lambda a,b: a+b, hedged_triple)
    
    return hedged[:-1]
```

## 23. ★ Zipf's Law

Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f × r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

### a. Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?

```python
def zipf_plot(word_list, n = None, s = None):
    """
    Takes a word list (e.g. from a corpus) and a range of words
    Draws a scatterplot of word frequency (y-axis) against
    word rank (x-axis) on log scales to demonstrate Zipf's Law
    If `n` is specified, plots only the `n` most frequent words
    """
    
    fd = nltk.FreqDist(word_list)
    if n is None: n = len(fd)
    if not s is None: s = s[:n]
    
    plt.scatter(
        [r + 1 for r in range(n)],
        sorted(fd.values(), reverse=True)[:n],
        s = s
    )
    plt.xscale("log")
    plt.yscale("log")
    plt.show()
```

```python
zipf_plot(brown.words())
```

Given a that all the points should theoretically lie on a straight line, there is some diversion on both ends. The most frequent words are _more_ frequent than the expected from Zipf's law, and the least frequent words are _less_ frequent.

## 25. ★ Define a function find_language() that takes a string as its argument, and returns a list of languages that have that string as a word. Use the udhr corpus and limit your searches to files in the Latin-1 encoding.

```python
def find_lang(string):
    "Searches for a string in UDHR corpus. Latin-1 encoding only."
    
    latin1 = [lang for lang in udhr.fileids() if lang.endswith("-Latin1")]
    return [
        lang
        for lang in latin1
        if string in udhr.words(lang)
    ]

find_lang("und")
```

## 28.     ★ Use one of the predefined similarity measures to score the similarity of each of the following pairs of words. Rank the pairs in order of decreasing similarity. How close is your ranking to the order given here, an order that was established experimentally by (Miller & Charles, 1998): 

car-automobile, gem-jewel, journey-voyage, boy-lad, coast-shore, asylum-madhouse, magician-wizard, midday-noon, furnace-stove, food-fruit, bird-cock, bird-crane, tool-implement, brother-monk, lad-brother, crane-implement, journey-car, monk-oracle, cemetery-woodland, food-rooster, coast-hill, forest-graveyard, shore-woodland, monk-slave, coast-forest, lad-wizard, chord-smile, glass-magician, rooster-voyage, noon-string.

```python
pairs_str = "car-automobile, gem-jewel, journey-voyage, boy-lad, coast-shore, asylum-madhouse, magician-wizard, midday-noon, furnace-stove, food-fruit, bird-cock, bird-crane, tool-implement, brother-monk, lad-brother, crane-implement, journey-car, monk-oracle, cemetery-woodland, food-rooster, coast-hill, forest-graveyard, shore-woodland, monk-slave, coast-forest, lad-wizard, chord-smile, glass-magician, rooster-voyage, noon-string"
pairs = [
    (p[:p.index("-")], p[p.index("-")+1:])
    for p in pairs_str.split(", ")
]

def pair_similarity_max(pair):
    "Takes a 2-tuple of strings and measures syntactic similarity."
    
    syn_0 = wordnet.synsets(pair[0], "n")
    syn_1 = wordnet.synsets(pair[1], "n")
    similarity = [s0.path_similarity(s1) for s0 in syn_0 for s1 in syn_1]
    
    return max(similarity)

similarity_dict = {pair: pair_similarity_max(pair) for pair in pairs}
```

```python
sorted(similarity_dict.items(), key=lambda i: i[1], reverse=True)
```
