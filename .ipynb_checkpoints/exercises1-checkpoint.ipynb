## 19. ◑ What is the difference between the following two lines? Which one will give a larger value? Will this be the case for other texts?

```
>>> sorted(set(w.lower() for w in text1))
>>> sorted(w.lower() for w in set(text1))
```

- The second one will give a larger value, because `lower()` is called on the list resulting from `set(text1)`, which might contain words counted more than once due to different cases.

- In contrast, the first one converts the text to lowercase _before_ calling `set()`, which would make the words of the same case before extracting the distinct ones.

````python
print(
    len(sorted(set(w.lower() for w in text1))),
    len(sorted(w.lower() for w in set(text1)))
)
````

## 20. ◑ What is the difference between the following two tests: `w.isupper()` and `not w.islower()`?

Both should produce the same results except when given mixed case strings or strings with only puncuation symbols and no letters: `isupper()` will always be `False`, and `islower()` will also be `False` whose negation will always be `True`.

````python
print(
    "HELLO".isupper(),
    "hello".isupper(),
    "heLlO".isupper(),
    "HELLO...".isupper(),
    "hello...".isupper(),
    "^#$@$".isupper()
)
print(
    not "HELLO".islower(),
    not "hello".islower(),
    not "heLlO".islower(),
    not "HELLO...".islower(),
    not "hello...".islower(),
    not "^#$@$".isupper()
)
````

## 22. ◑ Find all the four-letter words in the Chat Corpus (text5). With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency.

````python
text5_four_letters = [w.lower() for w in text5 if len(w) == 4]
fdist5_four_letters = FreqDist(text5_four_letters)

print(fdist5_four_letters.most_common(10))
fdist5_four_letters.plot(10)
````

##  28. ◑ Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

````python
def percent(word, text):
    
    n_word = len([w for w in text if w == word])
    n_text = len(text)
    ratio  = n_word / n_text
    
    return str(round(100 * ratio, 4)) + "%"

percent("the", text1)
````
